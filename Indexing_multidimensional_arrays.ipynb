# Printing a Twodimensional array

```python
bar = np.array([
    [5,10,15,20],
    [25,30,35,40],
    [45,50,55,60]
])
print(bar)
```

# Slicing for Row

```python
bar[:1] # it tells array to give me every thing upto excluding "1"
```

```python
bar[1:3,[-2,-1]]
'''It tells array to give me the valus
 between rows one and three (three exclusive)and the second
to last column and last column'''
```

# Modifying the Twodimensional array

```python
bar[0,0]=-1
print (bar)  # hear we replaced first element with -1
```

```python
bar[1]=bar[2]
print(bar) # hear we replaced second row with third row
```

```python
bar [[0,1,2],[0,1,2]]=[0,0,0]
print (bar)
```
