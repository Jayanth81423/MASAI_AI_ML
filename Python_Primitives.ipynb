```python
# String (str)
x = "Hello, "

# Stringsconcatenation:
print(x + "world!")

# String Slicing
s="hello world"
print(s[2:4])
print(s[:5])
print(s[6:])

#String Length and Membership
s="hello world"
print(len(s))
print('w' in s) 
print('z' not in s)

#String Formatting
age = 25
print(f"{name} is {age} years old.")
```

```python
#types of formtting

#% Formatting
name = "jayanth"
age = 25    
print("%s is %d years old." % (name, age))

#str.format() Method
name = "jayanth"
age = 25
print("{} is {} years old.".format(name, age))

#f-Strings (Python 3.6+)
name = "jayanth"
age = 25
print(f"{name} is {age} years old.")
```

```python
#Common String Methods
s = " hello world "
print(s.upper())
print(s.lower())    
print(s.strip())
print(s.replace("world", "there"))  
print(s.split())    
print(s.find("world"))
print(s.startswith(" hello"))
print(s.endswith("world "))
print(s.count("o"))
print(s.capitalize())
print(s.title())
print(s.isalpha())
print(s.isdigit())
print(s.isalnum())
print(s.center(20))
print(s.ljust(20))
print(s.rjust(20))
print(s.swapcase())
print(s.format())
print(s.index("world"))
print(s.partition(" "))
print(s.rpartition(" "))
print(s.zfill(20))
print(s.encode())   
print(s.expandtabs())
print(s.translate(str.maketrans("h", "H")))
print(s.removeprefix(" hello"))
print(s.removesuffix("world "))
print(s.casefold())
print(s.islower())
print(s.isupper())
print(s.istitle())          
print(s.splitlines())
print(s.rsplit())
print(s.join(["Hello", "there"]))
print(s.translate(str.maketrans({"h": "H", "w": "W"})))
print(s.format_map({"greeting": "Hello", "place": "world"}))
print(s.encode("utf-8"))
print(s.expandtabs(4))
print(s.translate(str.maketrans("", "", "aeiou")))
```

```python
 #Logical Operators
 x, y = 5, 10
print((x > 0) and (y > 0))   # True
print((x < 0) or (y > 0))    # True
print(not(x == 5))           # False
```

```python
#Relational (Comparison) Operators
a, b = 10, 20
print(a < b, a == b, a != b)
```

```python
#Bitwise Operators
a, b = 5, 3
print(a & b, a | b, a ^ b, ~a, a << 1, a >> 1)
```

Bitwise Operators
&---AND
|---OR
^---XOR
~---NOT
<<--Left Shift
>>--Right Shift


```python
# Membership Operators
list = [1, 2, 3, 4, 5]
print(3 in list)   # True
print(6 in list)   # False
print(7 not in list)  # True
```

```python
#Identity Operators
a = 5
b = 5
c = [1, 2, 3]
d = [1, 2, 3]
print(a is b)   # True
print(a is not b)   # False
print(c is d)   # False (different objects)
print(c is not d)   # True
```
