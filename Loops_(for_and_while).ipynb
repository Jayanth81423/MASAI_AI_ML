# Type 2: Looping Through Lists

```python
#Going Through Each Item:
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(f"I like {fruit}")
```

```python

for fruit in fruits: #Loop Variable = The current item
    print(fruit)   # 'fruit' is the loop variable
```

```python
# With Index Numbers:
fruits = ["apple", "banana", "cherry"]

for i in range(len(fruits)):
    print(f"{i+1}. {fruits[i]}")
```

```python
# Better Way - Using enumerate():
students = ["Alice", "Bob", "Charlie", "Diana"]
for i, student in enumerate(students, 1):
    print(f"Student {i}: {student}")
```

```python
# Start counting from 1:
fruits = ["apple", "banana", "cherry"]

for index, fruit in enumerate(fruits, 1):
    print(f"{index}. {fruit}")
```

# Type 4: Looping Through Multiple Lists Together

```python
# Guessing Game
import random

secret_number = random.randint(1, 10)
guess = 0

print("I'm thinking of a number between 1 and 10")

while guess != secret_number:
    guess = int(input("Your guess: "))

    if guess < secret_number:
        print("Too low! Try again")
    elif guess > secret_number:
        print("Too high! Try again")
    else:
        print("Correct! ðŸŽ‰")
```

```python
# Keep adding numbers until user says "done"
total = 0

while True:  # Loop forever until we break
    user_input = input("Enter a number (or 'done' to finish): ")

    if user_input == "done":
        break  # Exit the loop

    number = int(user_input)
    total += number

print(f"Total sum: {total}")

```

```python
#Countdown Timer
import time

seconds = int(input("Enter the time limit in seconds: "))

print("Countdown starting!")
while seconds > 0:
    print(f"{seconds} seconds remaining...")
    time.sleep()  # Wait 1 second
    seconds -= 1

print("Time's up! â°")
```

```python
prices = [20, 15, 30, 10, 25]
budget = 30
spent = 10

for price in prices:
    if spent + price > budget:
        print(f"Can't afford â‚¹{price} - budget exceeded!")
        break

    spent += price
    print(f"Bought item for â‚¹{price}. Spent: â‚¹{spent}")

print(f"Total spent: â‚¹{spent}")
```

```python
# Print only positive numbers
numbers = [5, -3, 8, -1, 10, -7, 3]

for num in numbers:
    if num < 0:
        continue  # Skip negative numbers

    print(num)

```

```python
grades = [85, 45, 92, 38, 20, 55, 88]
passing_grade = 50

print("Passing grades:")
for grade in grades:
    if grade < passing_grade:
        continue  # Skip failing grades

    print(grade)
```

```python
# break vs continue - Side by Side
print("=== Using break ===")
for i in range(1, 6):
    if i == 3:
        break  # Stop completely at 3
    print(i)
# Output: 1, 2

print("\n=== Using continue ===")
for i in range(1, 6):
    if i == 3:
        continue  # Skip only 3
    print(i)
# Output: 1, 2, 4, 5
```

```python
# break - Exit loop completely
for i in range(10):
    if i == 5:
        break  # Stops at 5

# continue - Skip to next iteration
for i in range(10):
    if i == 5:
        continue  # Skips 5, continues with 6
    print(i)

# pass - Do nothing (placeholder)
for i in range(10):
    if i == 5:
        pass  # Placeholder for future code
```
