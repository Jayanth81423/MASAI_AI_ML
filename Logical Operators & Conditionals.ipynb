Logical Operators & Conditionals

```python
# Logical Operators
# And: Both must be true
age = 20
nationality = "Indian"
if age >= 18 and nationality == "Indian":        
    print("Eligible to vote")

# OR: Only one needs to be True
is_raining = False
has_umbrella = True
print(is_raining or has_umbrella)  # Output: True

#Not: negates the value
is_sunny = True
print(not is_sunny)  # Output: False
```

Conditional Statements

```python
marks = 82

if marks >= 90:
    print("Grade A")   
elif marks >= 75:
    print("Grade B")      
elif marks >= 60:
    print("Grade C")   
else:       
    print("Grade D")      
```
