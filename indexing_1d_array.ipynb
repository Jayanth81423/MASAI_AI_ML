```python
num = np.array([10,20,30,40,50,60])
print(num)
```

# To modify an element in a specific index

```python
num[3]=80
print(num)
```

# Negative indexing
-> It starts from -1 and it represents last value of an array

```python
print(num[-1])
""" We are asking to print the value at -1 index
witch gives the last value of array num[] and output is 60"""
```

Getting the array elements by giving the list of indices

```python
num[[2,4,3]]
```

```python
num[np.zeros(shape=4,dtype='int64')]
```

#Slicing of an array
->arr[start index: end index : step size]


```python
num[0:5:2]
# Slicing an array num with a starting index of 0, ending index of 5 and with step of 2
```

```python
num[2:]  #num = np.array([10 20 30 80 50 60])
# [2:]->gives the valus from index '2'
```

```python
num[:2]  #num = np.array([10 20 30 80 50 60])
# [:2]-> gives the valus before index 2
```

```python
num[::2]  #num = np.array([10 20 30 80 50 60])
#[::2]-> gives all values in an array with step 2
```

# Updating a group list of indices in an array
->The list of values needs to be he same as list of indices

```python
num[[0,1,2,3]] = [100,200,300,400] #list of values it
print(num)             # needs to be same as list of indices
```

```python
# to assign one value to multiple indices, one value to multiple indices we use
num[[0,1]] = 500
print(num)
```
